```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc

#  1. Load Data

df = pd.read_csv('/content/Titanic-Dataset.csv')

print("Dataset loaded successfully.")
print(f"Dataset shape: {df.shape}")
print("\nFirst 5 rows of features:")
print(df.head())

#  2. Explore Data

print("\n Dataset Info")
df.info()

# Descriptive statistics
print("\n Descriptive Statistics ")
print(df.describe())

# Check for missing values
print("\n Missing Values ")
print(df.isnull().sum())

# Visualize target distribution
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Distribution of Target Variable (0: Not Survived, 1: Survived)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Not Survived', 'Survived'])
plt.show()

#  3. Prepare Data
# Drop irrelevant columns for modeling
# 'PassengerId', 'Name', 'Ticket', 'Cabin' are typically not used as features
df_cleaned = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)

# Handle missing 'Age' values by imputing with the mean
df_cleaned['Age'].fillna(df_cleaned['Age'].mean(), inplace=True)

# Handle missing 'Embarked' values by imputing with the mode
# There are only 2 missing values, so mode imputation is reasonable
df_cleaned['Embarked'].fillna(df_cleaned['Embarked'].mode()[0], inplace=True)

# Convert categorical features ('Sex', 'Embarked', 'Pclass') to numerical using one-hot encoding
# drop_first=True avoids multicollinearity by dropping the first category
df_processed = pd.get_dummies(df_cleaned, columns=['Sex', 'Embarked', 'Pclass'], drop_first=True)

# Define features (X) and target (y)
X = df_processed.drop('Survived', axis=1) # Features are all columns except 'Survived'
y = df_processed['Survived']              # Target is the 'Survived' column

# Split data into training and testing sets
# test_size is set to 0.3 as requested by the user
# random_state for reproducibility
# stratify=y ensures that the proportion of target variable is maintained in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"\nTraining features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
print(f"Training target shape: {y_train.shape}")
print(f"Testing target shape: {y_test.shape}")

#  4. Create and Train the Logistic Regression Model
# Initialize the Logistic Regression model
# 'liblinear' solver is good for small datasets and supports L1/L2 regularization
# max_iter increased to ensure convergence
model = LogisticRegression(solver='liblinear', random_state=42, max_iter=200)

# Train the model using the training data
print("\nTraining the Logistic Regression model")
model.fit(X_train, y_train)
print("Model training complete.")

#  5. Make Predictions
# Predict probabilities for the positive class (class 1 - Survived)
y_pred_proba = model.predict_proba(X_test)[:, 1]

# Predict class labels
y_pred = model.predict(X_test)

#  6. Evaluate the Model
print("\n Model Evaluation ")

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted Not Survived', 'Predicted Survived'],
            yticklabels=['Actual Not Survived', 'Actual Survived'])
plt.title('Confusion Matrix for Titanic Survival Prediction')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Classification Report (Precision, Recall, F1-score)
class_report = classification_report(y_test, y_pred, target_names=['Not Survived', 'Survived'])
print("\nClassification Report:")
print(class_report)

# ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Titanic Survival Prediction')
```
